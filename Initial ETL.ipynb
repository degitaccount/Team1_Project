```python
# Initial imports
import pandas as pd
# Load the yearly.csv datasets.
file_path2020 = "2020.csv"
happiness_2020_df = pd.read_csv(file_path2020)
happiness_2020_df.head(10)
year = 2020
happiness_2020_df["Year"] = year
happiness_2020_df
```

```python
output_file_path2020 = "new_2020.csv"
happiness_2020_df.to_csv(output_file_path2020, index=False)
```
